````python
import pandas as pd
````

````python
df = pd.read_excel('titanic_M1.xlsx')
df.head()
````

````python
df.dtypes
````

````python
missing_values = df.isnull().sum()
missing_values
````

````python
df['pclass'] = df['pclass'].fillna(2)
df.isnull().sum()
````

````python
df['sex'].unique()
````

````python
df['sex'] = df['sex'].replace('Hombre', 'male')
df['sex'] = df['sex'].replace('Mujer', 'female')
df['sex'].unique()
````

````python
df['age'].unique()
````

````python
df_ages_not_null = df[df['age'].notnull()]
average_age = df_ages_not_null['age'].mean()
average_age
````

````python
df['age'] = df['age'].fillna(average_age)
df.isnull().sum()
````

````python
df['embarked'] = df['embarked'].fillna('S')
df.isnull().sum()
````
